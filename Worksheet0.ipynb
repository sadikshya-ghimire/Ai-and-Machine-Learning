Create a Python program that converts between different units of measurement.

• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and

• Conversion Options:
1. Length:

      – Convert meters (m) to feet (ft).

      – Convert feet (ft) to meters (m).

2. Weight:

      – Convert kilograms (kg) to pounds (lbs).

      – Convert pounds (lbs) to kilograms (kg).

3. Volume:

      – Convert liters (L) to gallons (gal).

      – Convert gallons (gal) to liters (L).

```python
def conversion (a, choice):
  '''
  Convert a value between different units of measurement.

  Parameters:
  a (float): a float value
  choice (str): user input

  Returns:
  function returns the operation output.
  '''
  if (choice == "1"):
    meter_feet = a * 3.281
    feet_meter = a / 3.281

    return meter_feet, feet_meter

  elif (choice == "2"):
    kilo_pound = a * 2.2
    pound_kilo = a / 2.2

    return kilo_pound, pound_kilo

  elif (choice == "3"):
    liter_gallon = a * 0.264
    gallon_liter = a / 0.264

    return liter_gallon, gallon_liter




try:
    choice = input("Choose the type of conversion: \n"
                        "1. Length \n"
                        "2. Weight \n"
                        "3. Volume\n")
    value = float(input("Enter the number: "))

    if(choice == "1"):
      output1, output2 = conversion(value, choice)
      print(f"Meter -> Feet: {output1:.2f}, Feet -> Meter: {output2:.2f}")
    elif(choice == "2"):
      output1, output2 = conversion(value, choice)
      print(f"Kilogram -> Pound: {output1:.2f}, Pound -> Kilogram: {output2:.2f}")
    elif(choice == "3"):
      output1, output2 = conversion(value, choice)
      print(f"Liter -> Gallon: {output1:.2f}, Gallon -> Liter: {output2:.2f}")
    else:
      print("Invalid choice.")


except ValueError:
    print("Non-Numeric number is not excepted.")

```

• Requirements:

empty lists).

4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and

```python
def add(num):
  '''
  Sums the list number using iteration.

  Parameter:
  num[List]: list of userinput numbers.

  Returns:
  sum: operation output
  '''
  sum = 0
  for i in num:
    sum += i

  return sum

def average(output, num):
  '''
  Finds the average value of numbers in list

  Parameter:
  num[List]: list of userinput numbers.

  Returns:
  average: operation output
  '''
  average = output/len(num)

  return average


def maximum(num):
  '''
  Finds the maximum value from the list

  Parameter:
  num[List]: list of userinput numbers.

  Returns:
  max_num(float): maximum value
  '''
  max_num = num[0]
  for i in num:
    if i > max_num:
      max_num = i
  return max_num


def minimum(num):
  '''
  Finds the minumum value from the list

  Parameter:
  num[List]: list of userinput numbers.

  Returns:
  mini_num (float): minimum value
  '''
  mini_num = num[0]
  for i in num:
    if i < mini_num:
      mini_num = i

  return mini_num


try:
  choice = input("Choose an operation: \n"
                "1. Sum \n"
                "2. Average \n"
                "3. Maximum\n"
                "4. Minimum \n")
  user_input = input("Enter the list of numbers (separated by spaces)"). split()
  if not user_input:
        raise ValueError("Empty input! Please enter some numbers.")

  num = list(map(float, user_input))

  if (choice == "1"):
    output = add(num)
    print("Sum: ", output)

  elif (choice == "2"):
    output = add(num)
    output2 = average(output, num)
    print("Average: ", output2)

  elif (choice == "3"):
    output = maximum(num)
    print("Maximum: ", output)

  elif(choice == "4"):
    output = minimum(num)
    print("Minimum: ", output)

  else:
    print("Invalid Choice.")

except (ValueError, TypeError):
  print("Invalid Input.")
```

Write a Python function that returns a sublist from a given list, starting from a specified index and
ending at another specified index.

• Requirements:

– Define a function get sublist(lst, start, end) that takes a list lst, a starting index
start, and an ending index end as input and returns the sublist from start to end (inclusive).

– Example: For the input [1, 2, 3, 4, 5, 6] with start=2 and end=4, the output should
be [3, 4, 5].

```python
def get_sublist(lst, start, end):
  '''
  Returns a sublist from a given list starting from the specified start index
  and ending at the specified end index (inclusive).

  Parameters:
  lst (list): The original list from which to extract the sublist.
  start (int): The starting index of the sublist (inclusive).
  end (int): The ending index of the sublist (inclusive).

  Returns:
  sub_lst: A new list containing elements from index `start` to `end` of `lst`.
  '''

  sub_lst = []
  for i in range(start, end+1):
    sub_lst.append(lst[i])

  return sub_lst

try:
  user = input("Enter the list of numbers: "). split()
  if not user:
        raise ValueError("Empty input! Please enter some numbers.")
  lst = list(map(int, user))

  start = int(input("Enter the starting index: "))
  end = int(input("Enter the ending index: "))
  output = get_sublist(lst, start, end)
  print(output)

except (ValueError, IndexError) as e:
  print("Error: ", e)
```

1. Flatten a Nested List:

Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.

• Requirements:

– Define a function flatten(lst) that takes a nested list lst and returns a flattened version of the list.
– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(nested_lst):
  '''
    Flattens a 2D (nested) list into a single list.

    Parameters:
    nested_lst (list): A list containing other lists as elements.

    Returns:
    list: A single list containing all the elements from the nested lists.
  '''
  single_lst = []
  for i in range(len(nested_lst)):
    for j in range(len(nested_lst[i])):
      single_lst.append(nested_lst[i][j])

  return single_lst

nested_lst = [[1,2], [3,4],[5]]
output = flatten(nested_lst)
print(output)
```

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

– Define a function sum nested(lst) that takes a nested list lst and returns the sum of all the elements.

```python
def sum_nested(lst):
  '''
  Calculate the sum of all numbers in a nested list.

  Parameters:
  lst (list): A list that may contain integers and nested lists.

  Returns:
  int: The total sum of all numbers in the nested list.
  '''

  total = 0
  for i in lst:
    if isinstance(i, list):
      total += sum_nested(i)
    else:
      total += i
  return total


lst = [[1, 2], [3, [4, 5]], 6]
output = sum_nested(lst)
print(output)
```

Write a Python function that removes all occurrences of a specific element from a nested list.

• Requirements:

– Define a function remove element(lst, elem) that removes elem from lst and returns the modified list.

– Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
    '''
    Remove all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): A list that may contain nested lists.
    elem: The element to remove from the list.

    Returns:
    list: The modified list with all occurrences of elem removed.
    '''
    new_lst = []

    for i in lst:
        if isinstance(i, list):
            new_lst.append(remove_element(i, elem))
        else:
            if i != elem:
                new_lst.append(i)

    return new_lst


lst = [[1, 2], [3, 2], [4, 5]]
print(remove_element(lst, 2))
```

5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).

• Requirements:

– Define a function find max(lst) that takes a nested list lst and returns the maximum element.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    '''
    Find the maximum element in a nested list.

    Parameters:
    lst (list): A list that may contain integers and nested lists.

    Returns:
    int: The maximum value in the nested list.
    '''

    max_val = None

    for i in lst:
        if isinstance(i, list):
            value = find_max(i)
        else:
            value = i

        if max_val is None or value > max_val:
            max_val = value

    return max_val


lst = [[1, 2], [3, [4, 5]], 6]
print(find_max(lst))
```

6. Count Occurrences of an Element in a Nested List:

Write a Python function that counts how many times a specific element appears in a nested list.

• Requirements:

– Define a function count occurrences(lst, elem) that counts the occurrences of elem in the nested list lst.

– Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should be 3.

```python
def count_occurrences(lst, elem):
    '''
    Count how many times a specific element appears in a nested list.

    Parameters:
    lst (list): A list that may contain nested lists.
    elem: The element whose occurrences need to be counted.

    Returns:
    int: The total number of times elem appears in the nested list.

    '''

    count = 0

    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        else:
            if i == elem:
                count += 1

    return count


lst = [[1, 2], [2, 3], [2, 4]]
print(count_occurrences(lst, 2))
```

7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

• Requirements:

– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single flattened list.

– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be [1, 2, 3, 4, 5, 6, 7, 8].

```python
def flatten(lst):
    '''
    Flatten a deeply nested list into a single list.

    Parameters:
    lst (list): A list that may contain other lists at any depth.

    Returns:
    list: A single flattened list containing all elements.

    '''

    flat_list = []

    for i in lst:
        if isinstance(i, list):
            flat_list.extend(flatten(i))
        else:
            flat_list.append(i)

    return flat_list


lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(flatten(lst))
```

8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average nested(lst) that takes a nested list lst and returns the average of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    '''
    Calculate the average of all numbers in a nested list.

    Parameters:
    lst (list): A list that may contain integers and nested lists.

    Returns:
    float: The average of all numbers in the nested list.

    Example:
    >>> average_nested([[1, 2], [3, 4], [5, 6]])
    3.5
    '''

    total = 0
    count = 0

    for i in lst:
        if isinstance(i, list):
            sub_sum, sub_count = average_nested(i)
            total += sub_sum
            count += sub_count
        else:
            total += i
            count += 1

    return total, count


def calculate_average(lst):
    total, count = average_nested(lst)
    return total / count


lst = [[1, 2], [3, 4], [5, 6]]
print(calculate_average(lst))
```

```python

# Given matrices
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)

print("A inverse:")
print(A_inv)

print("\nA * A^-1 (Identity matrix):")
print(result)


# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nAB:")
print(AB)

print("\nBA:")
print(BA)


# 3. Prove (AB)^T = B^T A^T
left = (AB).T
right = np.dot(B.T, A.T)

print("\n(AB)^T:")
print(left)

print("\nB^T A^T:")
print(right)

# Solve Linear Equations AX = B
# System:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

A_matrix = np.array([[2,-3,1],
                     [1,-1,2],
                     [3,1,-1]])

B_matrix = np.array([-1,-3,9])

# Inverse method
A_inv = np.linalg.inv(A_matrix)

X = np.dot(A_inv, B_matrix)

print("\nSolution of equations (x, y, z):")
print(X)
```

4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import numpy as np
import time
import random

# Matrix size
n = 1000

# Using Python Lists
A = [[random.random() for _ in range(n)] for _ in range(n)]
B = [[random.random() for _ in range(n)] for _ in range(n)]

start = time.time()

# Initialize result matrix
result = [[0 for _ in range(n)] for _ in range(n)]

# Matrix multiplication
for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Time taken using Python lists:", end - start)

# Using NumPy Arrays
A_np = np.random.random((n, n))
B_np = np.random.random((n, n))

start = time.time()

result_np = np.dot(A_np, B_np)

end = time.time()

print("Time taken using NumPy arrays:", end - start)
```
